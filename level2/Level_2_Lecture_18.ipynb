```python
# 斐波那契数列：从第三项开始，每个数都是前两项的和
count_call_times = 0

def fibo(n):
    global count_call_times
    count_call_times +=1
    if n == 1 or n==2:
        return 1
    else:
        return fibo(n-2)+fibo(n-1)
print(f"调用斐波那契数列函数的结果为{fibo(40)}，次数为{count_call_times}")



```

```python
count_call_times = 0
memo = {}

def fibo(n):

    global count_call_times
    count_call_times +=1 # 用于计数
    if(n in memo):return memo[n]
    elif n == 1 or n==2:
        return 1
    else:
         ans = fibo(n-1) + fibo(n-2)
         memo[n] = ans
         return ans
print(f"调用斐波那契数列函数的结果为{fibo(40)}，次数为{count_call_times}")


```

```python
#迭代方式
def iterative(n):
    memo = [1,1]
    if(n > 2):
        for i in range(2,n):
            item = memo[i-2]+memo[i-1]
            memo.append(item)
    return memo[n-1]


print(iterative(14111))



```
