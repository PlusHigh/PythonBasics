### 2.格式化输出

###  一、%-formatting

最早 Python 的格式化字串（format string）是用與 C 語言類似的 %-formatting，透過% 運算符號，將在元組（tuple）中的一組變量依照指定的格式化方式輸出。如 %s（字串）、 %d（十進位整數）、 %f（浮點數）：

```python
"%s says  %d/%d is %f " %("曾滈",2,3,0.66667)
```

但這種輸出方式並不適合有很多變量的情況，容易搞錯也不易讀，如：

```python
"I am %s%s%s." % ("曾","庆","滈")
```

### 二、str.format()

2008 發布的 Python 2.6 開始有新的格式化字符函數 str.format()，透過 {} 和 format 來代替 %運算符號，如我們可以將上面例子的：

- s = 'I am %s %s. %s'

重寫為：

- s = 'I am {first_name} {middle_name}. {last_name}'

这样有什么好处？ 可读性更好了，more human readable.

然後透過 format 來帶入值：

- s.format(first_name='Monkey', middle_name='D', last_name='Luffy')

### 三、f-string

- 2016 發布的 Python 3.6 新增 f-string，可以解決 %-formatting 變量不易閱讀以及 str.format() 接變量後變超長的問題,下面是一个使用format的例子。

```python
import math
print(f'The value of pi is approximately {math.pi:.3f}.')
```

```python
s = 'I am {first_name} {middle_name}. {last_name}'
s.format(first_name='Monkey', middle_name='King', last_name='。')
```

####  这是第一种命名方式——给每个需要格式化的位置命名，然后在 format进行变量赋值

#### 另一种—— 使用 format 后面的元组索引位置格式

```python
print("1 乘以{0}等于{0},2 乘以{1}等于{2} ".format (3,4,8))
```

```python
# variables
first_name = "Monkey"
middle_name = "D"
last_name = "Luffy"
# str.format()
"I am {first_name} {middle_name}. {last_name}".format(
     first_name=first_name,
     middle_name=middle_name,
     last_name=last_name
)
# Output: "I am Monkey D. Luffy"
```

可用 f-string 改成如下格式，就可以將之前的變量帶入並格式化輸出

```python
# variables
first_name = "Monkey"
middle_name = "D"
last_name = "Luffy"
# f-string
f"I am {first_name} {middle_name}. {last_name}"
# Output: "I am Monkey D. Luffy"
```

这种方法也是目前主流的用于输出字符串的方法在javascipt中叫做模版字符串。

我们比较一下他们的可读性


```python
import math
print(f'The value of pi is approximately {math.pi:.3f}.')
```

###  3.split 以及join函数

- 这两个函数都需要用到分隔符
- 因为我们可能会需要拆分字符串，比如拆分一个人的名字和姓氏，或者处理CSV（逗号分隔文件），我们将csv文件读入内存后使用split函数对一行进行划分获取我们想要的字段，使用join将字符串连接起来并写回csv文件

####  小练习


##### Split the string into list of strings 将此字符串分割为列表

- Input : Geeks for Geeks
- Output : ['Geeks', 'for', 'Geeks']


##### Join the list of strings into a string based on delimiter ('-') 使用指定的字符 - 将列表中的元素链接为一个字符串

- Input :  ['Geeks', 'for', 'Geeks']
- Output : Geeks-for-Geeks

```python
# Python code
# to split and join given string

# input string
s = 'Geeks for Geeks'
# print the string after split method
print(s.split(" "))
# print the string after join method
print("-".join(s.split()))


# this code is contributed by gangarajula laxmi
```
